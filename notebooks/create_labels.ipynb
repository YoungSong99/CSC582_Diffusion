```python

import torch

def inspect_label_counts(loader):
    type_ids, color_ids, shape_ids = [], [], []

    for batch in loader:
        type_ids.append(batch["type_id"])
        color_ids.append(batch["color_id"])
        shape_ids.append(batch["shape_id"])

    type_ids = torch.cat(type_ids)
    color_ids = torch.cat(color_ids)
    shape_ids = torch.cat(shape_ids)

    print("num_types:", type_ids.max().item() + 1)
    print("num_colors:", color_ids.max().item() + 1)
    print("num_shapes:", shape_ids.max().item() + 1)

    print("unique type_ids:", torch.unique(type_ids).tolist())
    print("unique color_ids:", torch.unique(color_ids).tolist())
    print("unique shape_ids:", torch.unique(shape_ids).tolist())

inspect_label_counts(train_loader)
```

```python
import json
import pandas as pd

def make_mapping_from_dataframe(df, save_path):
    mapping = {
        "type_id_to_name": dict(
            sorted(zip(df["type_id"], df["type"]))
        ),
        "color_id_to_name": dict(
            sorted(zip(df["primary_color_id"], df["primary_color"]))
        ),
        "shape_id_to_name": dict(
            sorted(zip(df["shape_id"], df["shape"]))
        ),
    }

    mapping["num_types"] = max(mapping["type_id_to_name"].keys()) + 1
    mapping["num_colors"] = max(mapping["color_id_to_name"].keys()) + 1
    mapping["num_shapes"] = max(mapping["shape_id_to_name"].keys()) + 1

    with open(save_path, "w") as f:
        json.dump(mapping, f, indent=2)

    return mapping

mapping = make_mapping_from_dataframe(
    pd.read_csv("/home/young/projects/pokemon_diffusion/data/05_model_input/train_metadata.csv"),
    "/home/young/projects/pokemon_diffusion/data/05_model_input/label_mapping.json"
)
```
