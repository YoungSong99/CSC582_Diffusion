```python
import torch
from PIL import Image
import matplotlib.pyplot as plt
from torchvision import transforms
from diffusers import AutoencoderKL

device = "cuda" if torch.cuda.is_available() else "cpu"
dtype = torch.float16 if device == "cuda" else torch.float32

vae = AutoencoderKL.from_pretrained(
    "stabilityai/sd-vae-ft-mse",
    torch_dtype=dtype
).to(device)

vae.eval()
```

```python
from pokemon_diffusion.training.datamodule import create_dataloaders

params = catalog.load("params:vae")
train_metadata = catalog.load("train_metadata")
val_metadata = catalog.load("val_metadata")

train_loader, val_loader = create_dataloaders(
    train_metadata=train_metadata,
    val_metadata=val_metadata,
    img_size=params["data"]["img_size"],
    batch_size=500,
    normalization_range="minus_one_to_one",
    num_workers=params["data"].get("num_workers", 2),
)
```

```python
batch = next(iter(val_loader))
x = batch["image"] if isinstance(batch, dict) else batch[0]
x = x[:8].to(device, dtype=dtype)


x = torch.nn.functional.interpolate(
    x, size=(512, 512), mode="bilinear", align_corners=False
)

with torch.no_grad():
    latent = vae.encode(x).latent_dist.sample()
    latent = latent * vae.config.scaling_factor
    recon = vae.decode(latent / vae.config.scaling_factor).sample

orig_img = (x / 2 + 0.5).clamp(0, 1)
recon_img = (recon / 2 + 0.5).clamp(0, 1)
```

```python
from torchvision.utils import make_grid
import matplotlib.pyplot as plt

grid = make_grid(torch.cat([orig_img, recon_img], dim=0), nrow=8)

plt.figure(figsize=(12, 4))
plt.imshow(grid.permute(1, 2, 0).float().cpu())
plt.axis("off")
plt.show()
```
