```python
def normalize_names(pokedex: pd.DataFrame, db: pd.DataFrame):
    pokedex["_name"] = pokedex["name"].str.strip().str.lower()
    db["_name"] = (
    db["Pokemon"]
        .str.strip()
        .str.lower()
        .str.replace(r"\s+forme$", "", regex=True)   
        .str.replace(r"%", "", regex=False)          
        .str.strip()
    )

    return pokedex, db


normalize_names(pokedex,)
```

```python
def left_merge(pokedex, db):
    merged = db.merge(pokedex, on="_name", how="left")
    return merged

merged = left_merge(pokedex, db)
matched = set(merged["_name"])
pokedex_unmatched = set(pokedex["_name"]) - matched

pokedex_unmatched
```

```python
import re

def clean_pokemon_name(name):
    if pd.isna(name):
        return ""
    
    name = str(name).lower()
    
    stopwords = [
        'mega', 'galarian', 'alolan', 'hisuian', 'paldean', 
        'standard', 'mode', 'zen', 'normal', 'attack', 'defense', 
        'origin', 'therian', 'crowned', 'shield', 'sword', 'form', 'forme',
        'two-segment', 'ice face', 'zen mode', 'gigantamax'
    ]
    
    pattern = r'\b(' + '|'.join(stopwords) + r')\b'
    name = re.sub(pattern, '', name)
    
    name = re.sub(r'[^a-z0-9]', ' ', name) 
    name = ' '.join(name.split())
    
    return name
```

```python
import numpy as np
from rapidfuzz import process, fuzz

temp = merged.copy()
temp['_name_clean'] = temp['_name'].apply(clean_pokemon_name)

ref_df = temp.dropna(subset=['name', 'type1', 'primary_color', 'shape']).copy()

target_mask = temp[['type1', 'primary_color', 'shape']].isnull().any(axis=1)
target_indices = temp[target_mask].index

ref_names = ref_df['name'].tolist()

for idx in target_indices:
    query_name = str(temp.at[idx, '_name_clean'])
    
    match = process.extractOne(
        query_name, 
        ref_names, 
        scorer=fuzz.token_sort_ratio
    )
    
    if match:
        best_name, score, _ = match
        
        words_query = set(query_name.lower().split())
        words_best = set(best_name.lower().split())
        has_overlap = not words_query.isdisjoint(words_best)
        
        if score >= 30 or has_overlap:
            source_row = ref_df[ref_df['name'] == best_name].iloc[0]
            
            temp.at[idx, 'name'] = source_row['name']
            temp.at[idx, 'type1'] = source_row['type1']
            temp.at[idx, 'primary_color'] = source_row['primary_color']
            temp.at[idx, 'shape'] = source_row['shape']

print(temp.loc[target_indices, ['_name_clean', 'name', 'type1']])
```
