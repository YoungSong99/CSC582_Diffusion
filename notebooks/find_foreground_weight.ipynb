```python
from pokemon_diffusion.training.datamodule import create_dataloaders

params = catalog.load("params:ae")
train_metadata = catalog.load("train_metadata")
val_metadata = catalog.load("val_metadata")

_, val_loader = create_dataloaders(
    train_metadata=train_metadata,
    val_metadata=val_metadata,
    img_size=params["data"]["img_size"],
    batch_size=1000,
    normalization_range="minus_one_to_one",
    num_workers=params["data"].get("num_workers", 2),
)
```

```python
import torch
import matplotlib.pyplot as plt

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

batch = next(iter(val_loader))
x = batch["image"].to(device)

x_01 = (x + 1) / 2 if params["data"]["normalization_range"] == "minus_one_to_one" else x
brightness = x_01.mean(dim=1, keepdim=True)
fg_mask = (brightness < 0.99).float()
num_images = 8
fig, axes = plt.subplots(2, num_images, figsize=(16, 4))

for i in range(num_images):
    img = x_01[i].detach().cpu().permute(1, 2, 0).clamp(0, 1)
    mask = fg_mask[i, 0].detach().cpu()
    print("foreground ratio:", mask.mean().item())
    print("background ratio:", 1 - mask.mean().item())

    axes[0, i].imshow(img)
    axes[0, i].axis("off")
    axes[1, i].imshow(mask, cmap="gray")
    axes[1, i].axis("off")

axes[0, 0].set_title("Image")
axes[1, 0].set_title("Mask")
plt.show()
```

```python
import matplotlib.pyplot as plt

ratios = []

for i in range(499):
    img = x_01[i].detach().cpu().permute(1, 2, 0).clamp(0, 1)
    mask = fg_mask[i, 0].detach().cpu()
    foreground_ratio = mask.mean().item()
    background_ratio = 1 - foreground_ratio
    ratios.append(background_ratio/foreground_ratio)

plt.hist(ratios, bins=50)
plt.xlabel("Background to Foreground Ratio")
plt.ylabel("Frequency")
plt.title("Distribution of Background to Foreground Ratios")
plt.show()
```

```python
print("mean:", sum(ratios) / len(ratios))
print("median:", sorted(ratios)[len(ratios)//2])
```
